```python
import geopandas as gpd
from shapely.geometry import Point
import pandas as pd
import matplotlib.pyplot as plt
```

```python
point = Point(36.8219, -1.2921)
# Create a GeoDataFrame with the point using WGS84
gdf_wgs84 = gpd.GeoDataFrame({
    'city': ['Nairobi']},
    geometry=[point],
    crs='EPSG:4326'  # WGS84
)
```

```python
gdf_wgs84
```

```python
# Reproject to UTM Zone 37N (EPSG:32637)
gdf_utm = gdf_wgs84.to_crs('EPSG:32637')  # UTM zone 37N
```

```python
gdf_utm
```

```python
gdf_utm.crs
```

```python
# Plot the point in both coordinate systems
fix, ax = plt.subplots(1, 2, figsize=(20, 5))
gdf_wgs84.plot(ax=ax[0], color='blue')
ax[0].set_title('WGS84 Coordinate System')
ax[0].set_xlabel('Longitude')
ax[0].set_ylabel('Latitude')
gdf_utm.plot(ax=ax[1], color='red')
ax[1].set_title('UTM Zone 37N Coordinate System')
ax[1].set_xlabel('Easting')
ax[1].set_ylabel('Northing')
plt.show()
```
